```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('height-weight.csv')
df.head()
```

```python
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('WEIGHT')
plt.ylabel('HEIGHT')
```

```python
# NOW WE NEED TO CREATE A BEST FIT LINE 
```

```python
# STEPS
# 1.divide the dataset into dependent and indepenedent
# 2.Train Test Split
# 
```

```python
x=df[['Weight']]# independent features
y=df['Height']# target features
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Standadization
# -->gradient decent converges faster
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
# model being trained 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
plt.scatter(xtrain,ytrain,cmap='g')
plt.plot(xtrain,regressor.predict(xtrain))
```

```python
residula=ytest-regressor.predict(xtest)
residula
```

```python
import seaborn as sns
sns.displot(residula,kde=True)
```
